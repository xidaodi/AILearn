```python
from matplotlib import pyplot as plt
```

```python
x=range(1,8)
y=[17,17,18,15,11,11,13]
```

```python
#绘制折线；alpha表示透明度，0-1；linewidth表示像素值,marker表示折点标记
plt.plot(x,y,color="red",alpha=0.5,linestyle="--",linewidth=3,marker="o",markersize="20",markeredgecolor="green")
#展示
plt.show()
```

```python
#图片保存到本地
#设置图片的大小，figsize表示宽*高，dpi表示每个英寸包含多少像素
plt.figure(figsize=(20,8),dpi=80)
#绘制
plt.plot(x,y)
#保存
plt.savefig("./t1.png")
```

```python
from matplotlib import pyplot as plt
import random
x=range(2,26,2)
y=[random.randint(15,30) for i in x]
print(list(x))
print(y)
```

```python
plt.figure(figsize=(20,8),dpi=80)
#设置x轴的刻度
# plt.xticks(range(1,25))
x_tick_label=[f"{i}:00" for i in x]
#rotation表示倾斜程度
plt.xticks(x,x_tick_label,rotation=45)

#设置y轴的刻度
y_tick_label=[f"{i}℃" for i in range(min(y),max(y))]
plt.yticks(range(min(y),max(y)),y_tick_label)

plt.plot(x,y)
#设置标题
#设置中文字体
# from matplotlib import font_manager
# my_font=font_manager.FontProperties(fname="C:\Windows\Fonts")
# plt.title("piture",fontproperties=my_font)
plt.title("piture",size=18)
#x轴的标题
plt.xlabel("Time")
#y轴的标题
plt.ylabel("Temprature")
plt.show()
```

```python
#绘制两条折线
from matplotlib import pyplot as plt
import random
x=range(2,26,2)
y=[random.randint(15,30) for i in x]
y_1=[random.randint(1,15) for i in x]
print(list(x))
print(y)
print(y_1)
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y,color="red",label="y")
plt.plot(x,y_1,color="blue",label="y_1")
#绘制图例
plt.legend(loc="upper right")
#绘制网格
plt.grid(alpha=0.3)
plt.show()
```

```python
#如何处理空值,练习
t[3,3:]=np.nan
print(t)
print(t.shape)
```

```python
for i in range(t.shape[1]):
    temp_col=t[:,i]
    print(temp_col)
    nan_num=np.count_nonzero(temp_col!=temp_col)
    if nan_num!=0:
        temp_col_not_nan=temp_col[temp_col==temp_col]
        print(f"the temp_col_not_nan is {temp_col_not_nan}")
        temp_col[np.isnan(temp_col)]=np.mean(temp_col_not_nan)
        
print(t)        
```
